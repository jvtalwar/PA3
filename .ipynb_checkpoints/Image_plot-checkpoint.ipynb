```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np


from torchvision import utils
from basic_fcn import *
from dataloader2 import *
from utils import *
import torchvision
import torch.nn.functional as F
import torch.optim as optim
from torch.autograd import Variable
import time
```

```python
def data2img(torch_img):
    img = torch_img.permute(1, 2, 0).numpy()
    for i in range(3):
        img[:, :, i] -= np.min(img[:, :, i])
        img[:, :, i] /= np.max(img[:, :, i])
    return img    
```

```python
def plot_image_label(img, lbl):
    fig, ax = plt.subplots(ncols = 2, figsize = (20, 10))
    ax[0].imshow(img)
    ax[1].imshow(lbl)
    plt.tight_layout()
```

```python
dataset = CityScapesDataset(csv_file='train.csv')

init_sample = dataset.__getitem__(11)
init_img = data2img(init_sample[0])
init_lbl = init_sample[2]
plot_image_label(init_img, init_lbl)
```

```python
dataset_aug = CityScapesDataset(csv_file='train.csv', transforms = ['rotation'], resize_factor=0.5)

trans_sample = dataset_aug.__getitem__(11)
trans_img = data2img(trans_sample[0])
trans_lbl = trans_sample[2]
plot_image_label(trans_img, trans_lbl)
```
